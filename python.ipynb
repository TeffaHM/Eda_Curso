````python
# We have defined n as size the list.
# thousand_n contains thousand values equal to n-1.

n = 1000000
thousand_n = [n - 1] * 1000
````

````python
%%time

lst = [i for i in range(n)]
for x in thousand_n:
    x in lst
````

````python
%%time

s = {i for i in range(n)}
for x in thousand_n:
    x in s
````
